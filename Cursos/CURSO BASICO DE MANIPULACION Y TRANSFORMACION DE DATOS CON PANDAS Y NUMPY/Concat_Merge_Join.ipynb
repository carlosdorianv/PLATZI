##Merge

```python
izq = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']
 })
```

```python
der = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']
 })
```

```python
izq.merge(der, on='Key')
```

```python
izq = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']
 })
der = pd.DataFrame({'Key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']
 })
```

```python
izq.merge(der, left_on='Key', right_on='Key_2')
```

```python
izq = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']
 })
der = pd.DataFrame({'Key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']
 })
```

```python
der
```

```python
izq.merge(der, left_on='Key', right_on='Key_2',how='right')
```

## Join - Index Match

```python
izq = pd.DataFrame({
    'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2']
)

der = pd.DataFrame({'C':['C0','C1','C2',],
 'D':['D0','D1','D2',]},
 index=['k0','k2','k3']
)
```

```python
izq.join(der)
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='outer')
```

```python
izq.join(der, how='outer')
```
